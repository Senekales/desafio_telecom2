```python
#Ahora vamos analizar la correlación entre las variables categoricas
#para esto debemos transformarlas en númericas.

#Seleccionamos los nombres de las columnas del tipo "object" y creamos una lista con ellas

object_cols = datos.select_dtypes(include='object').columns.tolist()
print(object_cols)
```

```python
#Vamos a utilizar One-Hot Encoding ya que no hay un orden intrinseco en las variables y el número de
#categorias denro de cada columna como vimos en la parte previa es razonable, entre 2 y 4 valores unicos.

from sklearn.preprocessing import OneHotEncoder

# Identificamos las variables númericas
numerical_cols = datos.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Instanciamos OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Realizamos Fit sobre las variables categoricas y las guardamos en una variable
encoded_data = encoder.fit_transform(datos[object_cols])

# Creamos un nuevo DataFrame con el resultado del Fit de las variables categoricas.
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(object_cols), index=datos.index)

# Unimos el dataFrame códificado anterior y lo únimos con las columnas númericas originales.
datos_encoded = datos[numerical_cols].join(encoded_df)

# Mostramos el nuevo DataFrame, cantidad y tipo de datos
display(datos_encoded.head())
display(datos_encoded.info())

```

```python
#Ahora estamos en condiciones de ver la correlación entre las variables categoricas

matriz_correlacion_categoricas = encoded_df.corr()
display(matriz_correlacion_categoricas)
```

```python
#graficamos la correlación para entender mejor

plt.figure(figsize=(20, 16))

sns.heatmap(matriz_correlacion_categoricas, annot=False, cmap='coolwarm',fmt=".2f" )
plt.title('Mapa de Calor de la Matriz de Correlación variables Categoricas')
plt.show()
```

```python
# Al observar la matriz de correlación podemos ver la existencia de correlaciones absolutas postivas como negativas
# por lo tanto, podriamos eliminar las columnas que tiene esta correlación dejando solo una, que representara los datos
# eliminados, para evitar la colinearidad.

# identificamos las correlaciones absolutas (absolute value = 1.0)
correlaciones_absolutas_unitarias = matriz_correlacion_categoricas.unstack().sort_values(kind="quicksort", ascending=False)
correlaciones_absolutas_unitarias = correlaciones_absolutas_unitarias[correlaciones_absolutas_unitarias.abs() == 1.0]

# Eliminamos las auto-correlaciones de las correlaciones absolutas unitarias
correlaciones_absolutas_unitarias = correlaciones_absolutas_unitarias[correlaciones_absolutas_unitarias.index.get_level_values(0) != correlaciones_absolutas_unitarias.index.get_level_values(1)]

print("\nCorrelaciones con valor absoluto igual a 1.0 (excluyendo auto-correlaciones):")
display(correlaciones_absolutas_unitarias.head(100))
```

```python
#terminado el proceso de eliminación de las correlaciones preparamos los datos para
#comenzar a entrenar modelos.

from sklearn.model_selection import train_test_split

# Definir características (X) y variable objetivo (y) para datos_encoded
# La variable objetivo es 'Churn_Yes'
if 'Churn_Yes' in datos_encoded.columns:
    X = datos_encoded.drop('Churn_Yes', axis=1)
    y = datos_encoded['Churn_Yes']
else:
    print("Error: 'Churn_Yes' no encontrada en el DataFrame. Por favor, especifica la variable objetivo.")
    X = None
    y = None

if X is not None and y is not None:
    # Dividir los datos en conjuntos de entrenamiento y prueba (división 70/30)
    # 'random_state' asegura la reproducibilidad
    # 'stratify=y' asegura que la proporción de la variable objetivo sea la misma en ambos conjuntos
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

    print("Forma de X_train original:", X_train.shape)
    print("Forma de X_test original:", X_test.shape)
    print("Forma de y_train:", y_train.shape)
    print("Forma de y_test:", y_test.shape)
```

```python
#vamos a entrenar un modelo de regresión de línea de base
from sklearn.dummy import DummyRegressor  # Modelo de regresión de línea de base

# DummyRegressor predice la media de los datos de entrenamiento por defecto
baseline = DummyRegressor()
# Entrenar el modelo con los datos de entrenamiento
baseline.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred_dummy = baseline.predict(X_test)

def calcular_regresion(y_test, y_pred):
  rmse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error (y_test, y_pred)
  r2 = r2_score(y_test, y_pred)
  metricas = {
      'RMSE':round (rmse**(1/2),4),
      'MAE': round (mae,4),
      'R2': round (r2,4),
  }
  return metricas
```

```python
resultados_baseline = calcular_regresion (y_test, y_pred_dummy)
resultados_baseline
```

```python
#Ahora entrenamos un modelo RandomForestRegresor

from sklearn.ensemble import RandomForestRegressor

modelo = RandomForestRegressor(max_depth=5, random_state=42)
modelo.fit(X_train, y_train)
ypred = modelo.predict(X_test)
resultados_rf = calcular_regresion(y_test,ypred)
resultados_rf
```

```python
#visualizamos la predicción del error

from yellowbrick.regressor import prediction_error

Visualizer = prediction_error(modelo, X_train, y_train, X_test, y_test)
```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# Lista de columnas a eliminar (basado en correlaciones absolutas unitarias y negativas absolutas unitarias)
columns_to_drop_after_split = [
    'internet.StreamingTV_No internet service',
    "internet.TechSupport_No internet service",
    'internet.DeviceProtection_No internet service',
    'internet.OnlineBackup_No internet service',
    'internet.OnlineSecurity_No internet service',
    'internet.InternetService_No',
    'phone.MultipleLines_No phone service',
    'Churn_No',
    'customer.gender_Female',
    'customer.Partner_No',
    'customer.Dependents_No',
    'phone.PhoneService_No',
    'account.PaperlessBilling_No'
]

# Eliminar las columnas especificadas de los conjuntos de entrenamiento y prueba
X_train = X_train.drop(columns=columns_to_drop_after_split, errors='ignore')
X_test = X_test.drop(columns=columns_to_drop_after_split, errors='ignore')

# Instanciar RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# Aplicar submuestreo al conjunto de entrenamiento
X_train_resampled, y_train_resampled = rus.fit_resample(X_train, y_train)

print("\nForma de X_train después de eliminar columnas y submuestreo:", X_train_resampled.shape)
print("Forma de y_train después de eliminar columnas y submuestreo:", y_train_resampled.shape)
print("Forma de X_test después de eliminar columnas:", X_test.shape)
print("\nDistribución de clases en y_train original:")
display(y_train.value_counts())
print("\nDistribución de clases en y_train después de submuestreo:")
display(y_train_resampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Instanciar el modelo de Regresión Logística
modelo_base = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' es bueno para conjuntos de datos pequeños y binarios

# Entrenar el modelo con los datos de entrenamiento balanceados y con columnas eliminadas
modelo_base.fit(X_train_resampled, y_train_resampled)

print("Modelo de Regresión Logística entrenado con datos balanceados y columnas eliminadas.")

# Ahora podrías evaluar el modelo en el conjunto de prueba (X_test, y_test)
# Podemos hacer esto en el siguiente paso si lo deseas.
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Realizar predicciones en el conjunto de prueba
y_pred = modelo_base.predict(X_test)

# Calcular y mostrar métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy:.4f}")

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusión:")
display(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Instanciar el modelo de Bosque Aleatorio
# Puedes ajustar los parámetros (ej. n_estimators, max_depth, random_state)
modelo_rf = RandomForestClassifier(random_state=42)

# Entrenar el modelo con los datos de entrenamiento balanceados y con columnas eliminadas
modelo_rf.fit(X_train_resampled, y_train_resampled)

print("Modelo de Bosque Aleatorio entrenado con datos balanceados y columnas eliminadas.")

# Ahora podrías evaluar el modelo en el conjunto de prueba (X_test, y_test)
# Podemos hacer esto en el siguiente paso si lo deseas.
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Realizar predicciones en el conjunto de prueba
y_pred_rf = modelo_rf.predict(X_test)

# Calcular y mostrar métricas de evaluación
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Precisión del modelo de Bosque Aleatorio: {accuracy_rf:.4f}")

print("\nReporte de Clasificación del Bosque Aleatorio:")
print(classification_report(y_test, y_pred_rf))

print("\nMatriz de Confusión del Bosque Aleatorio:")
display(confusion_matrix(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV

# Definir la cuadrícula de hiperparámetros a buscar
param_grid = {
    'n_estimators': [100, 200, 300],  # Número de árboles en el bosque
    'max_depth': [None, 10, 20],  # Profundidad máxima de los árboles
    'min_samples_split': [2, 5, 10], # Número mínimo de muestras requeridas para dividir un nodo interno
    'min_samples_leaf': [1, 2, 4]   # Número mínimo de muestras requeridas en un nodo hoja
}

# Instanciar GridSearchCV
# cv=5 para validación cruzada con 5 folds
# scoring='recall' es a menudo una buena métrica para optimizar en casos de desbalance de clases
# n_jobs=-1 usa todos los procesadores disponibles
grid_search = GridSearchCV(estimator=modelo_rf, param_grid=param_grid, cv=5, scoring='recall', n_jobs=-1)

# Ajustar GridSearchCV a los datos de entrenamiento balanceados
grid_search.fit(X_train_resampled, y_train_resampled)

# Mostrar los mejores hiperparámetros encontrados y el mejor score
print("Mejores hiperparámetros encontrados:", grid_search.best_params_)
print("Mejor recall (validación cruzada):", grid_search.best_score_)
```

```python
# Obtener los mejores hiperparámetros encontrados por GridSearchCV
best_params = grid_search.best_params_

# Instanciar un nuevo modelo de Bosque Aleatorio con los mejores hiperparámetros
modelo_rf_optimizado = RandomForestClassifier(random_state=42, **best_params)

# Entrenar el modelo optimizado con los datos de entrenamiento balanceados y con columnas eliminadas
modelo_rf_optimizado.fit(X_train_resampled, y_train_resampled)

print("Modelo de Bosque Aleatorio optimizado entrenado con datos balanceados y columnas eliminadas.")

# Realizar predicciones en el conjunto de prueba con el modelo optimizado
y_pred_rf_optimizado = modelo_rf_optimizado.predict(X_test)

# Calcular y mostrar métricas de evaluación para el modelo optimizado
accuracy_rf_optimizado = accuracy_score(y_test, y_pred_rf_optimizado)
print(f"\nPrecisión del modelo de Bosque Aleatorio optimizado: {accuracy_rf_optimizado:.4f}")

print("\nReporte de Clasificación del Bosque Aleatorio optimizado:")
print(classification_report(y_test, y_pred_rf_optimizado))

print("\nMatriz de Confusión del Bosque Aleatorio optimizado:")
display(confusion_matrix(y_test, y_pred_rf_optimizado))
```

## Evaluación e Interpretación de Modelos para la Predicción de Renuncia de Clientes

En este análisis, hemos explorado tres modelos para predecir la renuncia de clientes: un modelo de línea de base (`DummyRegressor`), un modelo de Bosque Aleatorio sin optimizar y un modelo de Bosque Aleatorio optimizado mediante `GridSearchCV`. El objetivo principal es identificar a los clientes que tienen una alta probabilidad de abandonar la empresa (`Churn_Yes`) para poder implementar estrategias de retención efectivas.

A continuación, se presentan las métricas clave de evaluación para cada modelo en el conjunto de prueba:

| Modelo                        | Precisión | Recall (Clase 'Yes') | F1-Score (Clase 'Yes') | Matriz de Confusión (Verdaderos Positivos) |
|-------------------------------|-----------|----------------------|------------------------|--------------------------------------------|
| Regresión Logística (Base)    | {{resultados_baseline['R2']}}    | 0.80                 | 0.62                   | 446                                        |
| Bosque Aleatorio (Sin optimizar) | {{round(modelo_rf.score(X_test, y_test), 4)}} | 0.78                 | 0.62                   | 437                                        |
| Bosque Aleatorio (Optimizado) | {{round(modelo_rf_optimizado.score(X_test, y_test), 4)}} | 0.81                 | 0.63                   | 455                                        |

**Interpretación de los Resultados:**

1.  **Modelo de Línea de Base (`DummyRegressor`):** Este modelo simple predice la media de la variable objetivo (`Churn_Yes`). Como se esperaba para un problema de clasificación binaria desbalanceado, su desempeño es bajo, con un R2 cercano a 0 y un recall moderado para la clase positiva, reflejando que simplemente predecir la media no es efectivo para identificar a los que renuncian.

2.  **Modelo de Bosque Aleatorio (Sin optimizar):** Este modelo muestra una mejora significativa en comparación con la línea de base. La precisión y el F1-score son más altos, y el recall para la clase "Yes" (clientes que renuncian) es de 0.78. Esto significa que el modelo es capaz de identificar correctamente al 78% de los clientes que efectivamente abandonaron. La matriz de confusión muestra que predijo correctamente a 437 clientes que renunciaron.

3.  **Modelo de Bosque Aleatorio (Optimizado):** Después de la optimización de hiperparámetros con `GridSearchCV` (buscando maximizar el recall), este modelo presenta las métricas más favorables para nuestro objetivo. Tiene la precisión más alta (0.7512) y, lo más importante, el **recall más alto para la clase "Yes" (0.81)**. Esto indica que el modelo optimizado es el más efectivo para identificar a los clientes que renuncian, logrando predecir correctamente a 455 de ellos (según la matriz de confusión). Aunque el F1-score tuvo una mejora modesta, el enfoque en maximizar el recall fue exitoso.

**Conclusión:**

Considerando la importancia de identificar a los clientes propensos a renunciar para implementar acciones de retención, el **Modelo de Bosque Aleatorio Optimizado** es la mejor opción entre los modelos evaluados. Su mayor recall en la clase "Yes" (0.81) lo hace más valioso para el negocio, ya que reduce la cantidad de falsos negativos (clientes que renuncian pero que el modelo predice que no lo harán), permitiendo así una intervención proactiva con un mayor número de clientes en riesgo. Aunque la precisión general es importante, en este contexto, minimizar los falsos negativos es crucial para la estrategia de retención de clientes.
